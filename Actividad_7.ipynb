# Ejercicio 1:
## Utilizando los dos métodos (secante y bisección), halle una raíz a la función f(x) = x^2 – 4.9x + 5.7cos(0.01x).

# Método de la secante.

```python
import matplotlib.pyplot as plt
import numpy as np
import math

def f(x):
    return x**2 - 4.9*x + 5.7*np.cos(0.01*x)

def secante(x0, x1, tol, max_iter):
    iteraciones = 0
    while iteraciones < max_iter:
        fx0 = f(x0)
        fx1 = f(x1)

        if abs(fx1 - fx0) < tol:
            print(f"Convergencia alcanzada. Resultado final: {x1}")
            return x1

        # Evitar la división por cero
        if fx1 == fx0:
            print("División por cero evitada. No se puede continuar.")
            return None

        x2 = x1 - (fx1 * (x1 - x0)) / (fx1 - fx0)

        print(f"Iteración {iteraciones + 1}: Xn+1 = {x2}    Xn = {x1}   Xn-1 = {x0}")

        x0, x1 = x1, x2
        iteraciones += 1

    print("Iteraciones máximas alcanzadas. Posible no convergencia.")
    return None

# Prueba del método de la secante
valor_inicial_1 = 1.0
valor_inicial_2 = 2.0
tolerancia_secante = 1e-6
max_iteraciones_secante = 100

resultado_secante = secante(valor_inicial_1, valor_inicial_2, tolerancia_secante, max_iteraciones_secante)

if resultado_secante is not None:
    print(f"Resultado final: {resultado_secante}")
else:
    print("No se alcanzó la convergencia en el número máximo de iteraciones.")

# Crear un rango de valores de x
x_vals = np.linspace(0.1, 4, 100)

# Calcular los valores de y correspondientes
y_vals = f(x_vals)

# Crear el gráfico con tamaño personalizado
plt.figure(figsize=(6, 4))

# Ploteo de la función
plt.plot(x_vals, y_vals, label='f(x) = x^2 – 4.9x + 5.7cos(0.01x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de la función f(x) = x^2 – 4.9x + 5.7cos(0.01x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# Ploteo de la raíz aproximada con línea vertical
if resultado_secante is not None:
    plt.plot(resultado_secante, 0, 'ro', label='Raíz Aproximada')
    plt.axvline(resultado_secante, color='red', linestyle='--', label='Línea Vertical - Raíz Aproximada')

# Mostrar el gráfico
plt.legend()
plt.show()
```

# Método de bisección

```python
import numpy as np
import math
def funcion(x):
    return x**2 - 4.9*x + 5.7*np.cos(0.01*x)

def biseccion(xi, xd, tolerancia):
    iteraciones = 0
    while True:
        xm = (xi + xd) / 2
        fxm = funcion(xm)
        print(f"Iteración {iteraciones + 1}: Xm = {xm}    Xi = {xi}   Xd = {xd}")
        if abs(fxm) < tolerancia:
            break

        if funcion(xi) * fxm > 0:
            xi = xm
        else:
            xd = xm

        iteraciones += 1

    return xm, iteraciones

# Definir intervalo inicial [xi, xd] y tolerancia
xi = 2.5
xd = 3.5
tolerancia = 1e-6

# Calcular el número de iteraciones usando la fórmula n = (ln(a) - ln(e))/ln(2)
num_iteraciones = (math.log(xd - xi) - math.log(tolerancia)) / math.log(2)

# Aplicar el método de bisección
raiz, _ = biseccion(xi, xd, tolerancia)

# Mostrar resultados
print(f"Raíz aproximada: {raiz}")
print(f"Número de iteraciones estimado: {num_iteraciones}")

# Crear un rango de valores de x
x_vals = np.linspace(0.1, 4, 100)

# Calcular los valores de y correspondientes
y_vals = funcion(x_vals)

# Crear el gráfico con tamaño personalizado
plt.figure(figsize=(6, 4))

# Ploteo de la función
plt.plot(x_vals, y_vals, label='f(x) = x^2 – 4.9x + 5.7cos(0.01x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de la función f(x) = x^2 – 4.9x + 5.7cos(0.01x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)

plt.plot(raiz, 0, 'ro', label='Raíz Aproximada')
plt.axvline(raiz, color='red', linestyle='--', label='Línea Vertical - Raíz Aproximada')

# Mostrar el gráfico
plt.legend()
plt.show()
```
