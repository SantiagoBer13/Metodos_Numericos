# **Metodo de Newton Raphson**


```python
import matplotlib.pyplot as plt
import numpy as np

# Definir la función
def funcion(x):
    return x**5 - 6

# Definir la derivada de la función
def derivada_funcion(x):
    return 5 * x**4

# Implementar el método de Newton-Raphson
def newton_raphson(func, func_derivada, x0, tolerancia=1e-10, max_iter=10):
    """
    Encuentra una raíz de la función 'func' utilizando el método de Newton-Raphson.

    Args:
        func (function): La función para la cual queremos encontrar una raíz.
        func_derivada (function): La derivada de la función 'func'.
        x0 (float): Aproximación inicial.
        tolerancia (float, optional): Tolerancia para la convergencia. El valor predeterminado es 1e-10.
        max_iter (int, optional): Número máximo de iteraciones permitidas. El valor predeterminado es 1000.

    Returns:
        float: Una estimación de la raíz de 'func'.
    """
    iteracion = 0
    x = x0
    print(f"Iteración {iteracion}:")
    print(f"x{iteracion}: {x}")
    print(f"f(x{iteracion}): {func(x)}")
    print(f"f'(x{iteracion}): {func_derivada(x)}")
    print()
    while iteracion < max_iter:
        fx = func(x)
        fpx = func_derivada(x)

        if abs(fx) < tolerancia:
            print(f"Convergencia alcanzada después de {iteracion} iteraciones.")
            return x

        x = x - fx / fpx
        iteracion += 1

        print(f"Iteración {iteracion}:")
        print(f"x{iteracion}: {x}")
        print(f"f(x{iteracion}): {fx}")
        print(f"f'(x{iteracion}): {fpx}")
        print()

    raise Exception(f"No se pudo encontrar una raíz después de {max_iter} iteraciones")

# Encontrar la raíz usando el método de Newton-Raphson
raiz_aproximada = newton_raphson(funcion, derivada_funcion, x0=2)
print(f"La raíz aproximada es: {raiz_aproximada}")

# Crear un rango de valores de x
x_vals = np.linspace(-2, 2, 100)  # Genera 100 puntos en el rango de -2 a 2

# Calcular los valores de y correspondientes
y_vals = funcion(x_vals)

# Crear el gráfico
plt.plot(x_vals, y_vals, label='f(x) = x^5 - 6')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de la función f(x) = x^5 - 6')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

# Agregar la raíz aproximada al gráfico
plt.plot(raiz_aproximada, 0, 'ro', label='Raíz Aproximada')

# Mostrar el gráfico
plt.legend()
plt.show()
```

# **Metodo** (Punto fijo)


```python
import math

def f(x):
    return x**3 - 3

def g(x):
    return 3 ** (1/3)

def punto_fijo(x0, tol, max_iter):
    """
    x0: Valor inicial
    tol: Tolerancia para la convergencia
    max_iter: Número máximo de iteraciones
    """
    iteraciones = 0
    while iteraciones < max_iter:
        x1 = g(x0)
        distancia = abs(x1 - x0)
        print(f"Iteración {iteraciones + 1}: x = {x1}, Distancia = {distancia}")
        if distancia < tol:
            print("Convergencia alcanzada.")
            return x1
        x0 = x1
        iteraciones += 1

    print("Iteraciones máximas alcanzadas. Posible no convergencia.")
    return None

# Prueba del método de punto fijo
valor_inicial = 1.0  # Puedes ajustar este valor
tolerancia = 1e-6
max_iteraciones = 100

resultado = punto_fijo(valor_inicial, tolerancia, max_iteraciones)

if resultado is not None:
    print(f"Resultado final: {resultado}")
else:
    print("No se alcanzó la convergencia en el número máximo de iteraciones.")
```

# Metodo de la Secante:

```python
import matplotlib.pyplot as plt
import numpy as np
import math

def f(x):
    return x**2 - 6.8*x**1.01 + 9

def secante(x0, x1, tol, max_iter):
    iteraciones = 0
    while iteraciones < max_iter:
        fx0 = f(x0)
        fx1 = f(x1)

        if abs(fx1 - fx0) < tol:
            print(f"Convergencia alcanzada. Resultado final: {x1}")
            return x1

        # Evitar la división por cero
        if fx1 == fx0:
            print("División por cero evitada. No se puede continuar.")
            return None

        x2 = x1 - (fx1 * (x1 - x0)) / (fx1 - fx0)

        print(f"Iteración {iteraciones + 1}: Xn+1 = {x2}    Xn = {x1}   Xn-1 = {x0}")

        x0, x1 = x1, x2
        iteraciones += 1

    print("Iteraciones máximas alcanzadas. Posible no convergencia.")
    return None

# Prueba del método de la secante
valor_inicial_1 = 1.0
valor_inicial_2 = 2.0
tolerancia_secante = 1e-6
max_iteraciones_secante = 100

resultado_secante = secante(valor_inicial_1, valor_inicial_2, tolerancia_secante, max_iteraciones_secante)

if resultado_secante is not None:
    print(f"Resultado final: {resultado_secante}")
else:
    print("No se alcanzó la convergencia en el número máximo de iteraciones.")

# Crear un rango de valores de x
x_vals = np.linspace(0.1, 4, 100)

# Calcular los valores de y correspondientes
y_vals = f(x_vals)

# Crear el gráfico con tamaño personalizado
plt.figure(figsize=(6, 4))

# Ploteo de la función
plt.plot(x_vals, y_vals, label='f(x) = x^2 – 4.9x + 5.7cos(0.01x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de la función f(x) = x^2 – 4.9x + 5.7cos(0.01x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# Ploteo de la raíz aproximada con línea vertical
if resultado_secante is not None:
    plt.plot(resultado_secante, 0, 'ro', label='Raíz Aproximada')
    plt.axvline(resultado_secante, color='red', linestyle='--', label='Línea Vertical - Raíz Aproximada')

# Mostrar el gráfico
plt.legend()
plt.show()
```

# Metodo de posicion falsa.

```python
import numpy as np

def f(x):
    return x**2 - 4.9*x + 5.7*np.cos(0.01*x)

def posicion_falsa(x_d, x_i, tol, max_iter):
    iteraciones = 0

    while iteraciones < max_iter:
        f_x_d = f(x_d)
        f_x_i = f(x_i)

        if f_x_d * f_x_i > 0:
            print("Error: Los valores iniciales no tienen signos opuestos.")
            return None

        # Fórmula de la posición falsa
        x_m = x_i - (f_x_i * (x_d - x_i)) / (f_x_d - f_x_i)

        f_x_m = f(x_m)

        if abs(f_x_m) < tol:
            print(f"Convergencia alcanzada. Resultado final: {x_m}")
            return x_m

        print(f"Iteración {iteraciones + 1}: x_m = {x_m}")

        # Actualizar los valores según el cambio de signo
        if f_x_m * f_x_d < 0:
            x_i = x_m
        else:
            x_d = x_m

        iteraciones += 1

    print("Iteraciones máximas alcanzadas. Posible no convergencia.")
    return None

# Prueba del método de posición falsa
valor_inicial_d = -1.0
valor_inicial_i = 2.0
tolerancia_posicion_falsa = 1e-6
max_iteraciones_posicion_falsa = 100

resultado_posicion_falsa = posicion_falsa(valor_inicial_d, valor_inicial_i, tolerancia_posicion_falsa, max_iteraciones_posicion_falsa)

if resultado_posicion_falsa is not None:
    print(f"Resultado final: {resultado_posicion_falsa}")
else:
    print("No se alcanzó la convergencia en el número máximo de iteraciones.")
```

# Metodo de biseccion

```python
import numpy as np
import math
def funcion(x):
    return x**2 - 6.8*x**1.01 + 9

def biseccion(xi, xd, tolerancia):
    iteraciones = 0
    while True:
        xm = (xi + xd) / 2
        fxm = funcion(xm)
        print(f"Iteración {iteraciones + 1}: Xm = {xm}    Xi = {xi}   Xd = {xd}")
        if abs(fxm) < tolerancia:
            break

        if funcion(xi) * fxm > 0:
            xi = xm
        else:
            xd = xm

        iteraciones += 1

    return xm, iteraciones

# Definir intervalo inicial [xi, xd] y tolerancia
xi = 5
xd = 6
tolerancia = 1e-6

# Calcular el número de iteraciones usando la fórmula n = (ln(a) - ln(e))/ln(2)
num_iteraciones = (math.log(xd - xi) - math.log(tolerancia)) / math.log(2)

# Aplicar el método de bisección
raiz, _ = biseccion(xi, xd, tolerancia)

# Mostrar resultados
print(f"Raíz aproximada: {raiz}")
print(f"Número de iteraciones estimado: {num_iteraciones}")

# Crear un rango de valores de x
x_vals = np.linspace(-1, 6, 100)

# Calcular los valores de y correspondientes
y_vals = funcion(x_vals)

# Crear el gráfico con tamaño personalizado
plt.figure(figsize=(6, 4))

# Ploteo de la función
plt.plot(x_vals, y_vals, label='f(x) = x^2 – 4.9x + 5.7cos(0.01x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de la función f(x) = x^2 – 4.9x + 5.7cos(0.01x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)

plt.plot(raiz, 0, 'ro', label='Raíz Aproximada')
plt.axvline(raiz, color='red', linestyle='--', label='Línea Vertical - Raíz Aproximada')

# Mostrar el gráfico
plt.legend()
plt.show()
```

# Metodo de Steffensen


```python
import numpy as np
def steffensen_method(G, X0, EPS, MAXIT):
    """
    Método de Steffensen para encontrar la raíz de la ecuación g(x) = x.

    Args:
    - G: Función de iteración.
    - X0: Valor inicial.
    - EPS: Criterio de convergencia.
    - MAXIT: Número máximo de iteraciones.

    Returns:
    - Raíz aproximada X o mensaje de falla.
    """

    I = 1

    while I <= MAXIT:
        X1 = G(X0)
        X2 = G(X1)

        numerator = (X1 - X0) ** 2
        denominator = X2 - 2 * X1 + X0

        # Evitar división por cero
        if denominator == 0:
            print("División por cero. El método no converge a una raíz.")
            return

        X = X0 - numerator / denominator

        if abs(X - X0) < EPS:
            print(f"La raíz aproximada es: {X}")
            return X

        I += 1
        X0 = X

    print("El método no converge a una raíz después de MAXIT iteraciones.")

# Ejemplo de uso:
def G(x):
    return x**2 - 4.9*x + 5.7 * np.cos(0.01*x)  # Define tu función g(x) aquí.

X0 = 1.0  # Define tu valor inicial aquí.
EPS = 1e-6  # Define tu criterio de convergencia aquí.
MAXIT = 100  # Define tu número máximo de iteraciones aquí.

steffensen_method(G, X0, EPS, MAXIT)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo dirigido para representar la red de amistades
G = nx.DiGraph()

# Agregar nodos (miembros del club)
members = ["Alfonso","Santiago", "Bob", "Charlie", "David", "Eva", "Frank"]
G.add_nodes_from(members)

# Establecer conexiones de amistad
friendships = [("Alfonso", "Bob"),("Alfonso", "Santiago"),("Alfonso", "Charlie"),("Alfonso", "David"),("Alfonso", "Eva"),("Alfonso", "Frank"),
 ("Santiago", "Bob"), ("Santiago", "Charlie"), ("Bob", "David"), ("Charlie", "David"),("Eva", "Frank")]

G.add_edges_from(friendships)

# Visualizar la red de amistades
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, font_weight='bold', node_color='lightblue', node_size=1000)
plt.title("Red de Amistades en la Materia Métodos Numéricos")
plt.show()

# Función para sugerir eventos entre amigos
def sugerir_eventos_entre_compañeros(graph, persona):
    amigos = list(graph.neighbors(persona))
    eventos_sugeridos = set()

    for amigo in amigos:
        eventos_amigo = set(graph.neighbors(amigo))
        eventos_sugeridos.update(eventos_amigo)

    eventos_sugeridos -= set(graph.neighbors(persona))  # Excluir eventos a los que la persona ya asistió
    return eventos_sugeridos

# Ejemplo de sugerencia de eventos para un miembro
miembro = "Alfonso"
eventos_sugeridos = sugerir_eventos_entre_compañeros(G, miembro)

print(f"Eventos sugeridos para {miembro}: {eventos_sugeridos}")
```
