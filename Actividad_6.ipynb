Ejercicio 1: Solución de la ecuación x = raíz quinta de 6 ​con el Método de Newton Raphson.

Pasos a seguir:

Primero, convertimos la ecuación a una forma igual a cero: x^5 − 6 = 0.

Definimos la función f(x) = x^5 − 6 y su derivada f′(x) = 5 x^4.

Implementamos el Método de Newton-Raphson en Python.

Utilizamos el método para encontrar una raíz aproximada.

```python
import matplotlib.pyplot as plt
import numpy as np

# Definir la función
def funcion(x):
    return x**5 - 6

# Definir la derivada de la función
def derivada_funcion(x):
    return 5 * x**4

# Implementar el método de Newton-Raphson
def newton_raphson(func, func_derivada, x0, tolerancia=1e-10, max_iter=10):
    """
    Encuentra una raíz de la función 'func' utilizando el método de Newton-Raphson.

    Args:
        func (function): La función para la cual queremos encontrar una raíz.
        func_derivada (function): La derivada de la función 'func'.
        x0 (float): Aproximación inicial.
        tolerancia (float, optional): Tolerancia para la convergencia. El valor predeterminado es 1e-10.
        max_iter (int, optional): Número máximo de iteraciones permitidas. El valor predeterminado es 1000.

    Returns:
        float: Una estimación de la raíz de 'func'.
    """
    iteracion = 0
    x = x0
    print(f"Iteración {iteracion}:")
    print(f"x{iteracion}: {x}")
    print(f"f(x{iteracion}): {func(x)}")
    print(f"f'(x{iteracion}): {func_derivada(x)}")
    print()
    while iteracion < max_iter:
        fx = func(x)
        fpx = func_derivada(x)

        if abs(fx) < tolerancia:
            print(f"Convergencia alcanzada después de {iteracion} iteraciones.")
            return x

        x = x - fx / fpx
        iteracion += 1

        print(f"Iteración {iteracion}:")
        print(f"x{iteracion}: {x}")
        print(f"f(x{iteracion}): {fx}")
        print(f"f'(x{iteracion}): {fpx}")
        print()

    raise Exception(f"No se pudo encontrar una raíz después de {max_iter} iteraciones")

# Encontrar la raíz usando el método de Newton-Raphson
raiz_aproximada = newton_raphson(funcion, derivada_funcion, x0=2)
print(f"La raíz aproximada es: {raiz_aproximada}")

# Crear un rango de valores de x
x_vals = np.linspace(-2, 2, 100)  # Genera 100 puntos en el rango de -2 a 2

# Calcular los valores de y correspondientes
y_vals = funcion(x_vals)

# Crear el gráfico
plt.plot(x_vals, y_vals, label='f(x) = x^5 - 6')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de la función f(x) = x^5 - 6')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

# Agregar la raíz aproximada al gráfico
plt.plot(raiz_aproximada, 0, 'ro', label='Raíz Aproximada')

# Mostrar el gráfico
plt.legend()
plt.show()
```
