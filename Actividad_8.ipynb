# Ejercicio 1

# Utilizando el método de la falsa posición, halle una raíz a la función

# f(x) = x^2 – 4.9x + 5.7cos(0.01x).



```python
import numpy as np

def f(x):
    return x**2 - 4.9*x + 5.7*np.cos(0.01*x)

def posicion_falsa(x_d, x_i, tol, max_iter):
    iteraciones = 0

    """
    Parámetros:
    - f: Función para la cual se busca una raíz.
    - xi: Extremo izquierdo del intervalo.
    - xd: Extremo derecho del intervalo.
    - tol: Tolerancia de error.
    - max_iter: Número máximo de iteraciones.

    Retorna:
    - Raíz aproximada de la función.
    """
    while iteraciones < max_iter:
        f_x_d = f(x_d)
        f_x_i = f(x_i)

        if f_x_d * f_x_i > 0:
            print("Error: Los valores iniciales no tienen signos opuestos.")
            return None

        # Fórmula de la posición falsa
        x_m = x_i - (f_x_i * (x_d - x_i)) / (f_x_d - f_x_i)

        f_x_m = f(x_m)

        if abs(f_x_m) < tol:
            print(f"Convergencia alcanzada. Resultado final: {x_m}")
            return x_m

        print(f"Iteración {iteraciones + 1}: x_d: {x_d}, x_i: {x_i}, x_m = {x_m}")

        # Actualizar los valores según el cambio de signo
        if f_x_m * f_x_d < 0:
            x_i = x_m
        else:
            x_d = x_m

        iteraciones += 1

    print("Iteraciones máximas alcanzadas. Posible no convergencia.")
    return None

# Prueba del método de posición falsa
valor_inicial_d = 1.0
valor_inicial_i = 2
tolerancia_posicion_falsa = 1e-6
max_iteraciones_posicion_falsa = 100

resultado_posicion_falsa = posicion_falsa(valor_inicial_d, valor_inicial_i, tolerancia_posicion_falsa, max_iteraciones_posicion_falsa)

if resultado_posicion_falsa is not None:
    print(f"Resultado final: {resultado_posicion_falsa}")
else:
    print("No se alcanzó la convergencia en el número máximo de iteraciones.")
```

```python
import matplotlib.pyplot as plt
# Crear un rango de valores de x
x_vals = np.linspace(1.75, 2, 100)

# Calcular los valores de y correspondientes
y_vals = f(x_vals)

# Crear el gráfico con tamaño personalizado
plt.figure(figsize=(6, 4))

# Ploteo de la función
plt.plot(x_vals, y_vals, label='f(x) = x^2 – 4.9x + 5.7cos(0.01x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de la función f(x) = x^2 – 4.9x + 5.7cos(0.01x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# Ploteo de la raíz aproximada con línea vertical
if resultado_posicion_falsa is not None:
    plt.plot(resultado_posicion_falsa, 0, 'ro', label='Raíz Aproximada')
    plt.axvline(resultado_posicion_falsa, color='red', linestyle='--', label='Línea Vertical - Raíz Aproximada')

# Mostrar el gráfico
plt.legend()
plt.show()
```
